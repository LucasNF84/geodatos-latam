# 🌍 Análisis de Países, Provincias y Localidades

Este notebook muestra un análisis exploratorio de los datos de países, provincias y localidades de América del Sur, utilizando los archivos descargados desde el repositorio `dr5hn/countries-states-cities-database`. Nos enfocamos en los siguientes países:

- 🇧🇷 Brasil
- 🇵🇾 Paraguay
- 🇺🇾 Uruguay
- 🇧🇴 Bolivia
- 🇨🇱 Chile
- 🇵🇪 Perú


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Leer CSVs
countries = pd.read_csv("countries.csv")
states = pd.read_csv("states.csv")
cities = pd.read_csv("cities.csv")
```

```python
# Filtrar solo los países deseados
target_codes = ['BR', 'PY', 'UY', 'BO', 'CL', 'PE']
countries = countries[countries['iso2'].isin(target_codes)]
states = states[states['country_code'].isin(target_codes)]
cities = cities[cities['country_code'].isin(target_codes)]
```

## 🗂️ Provincias por país

```python
prov_por_pais = states['country_code'].value_counts().rename_axis('country_code').reset_index(name='provincias')
sns.barplot(data=prov_por_pais, x='country_code', y='provincias', palette="Set2")
plt.title("Cantidad de provincias por país")
plt.xlabel("País")
plt.ylabel("Cantidad de provincias")
plt.show()
```

## 🏙️ Localidades por país

```python
local_por_pais = cities['country_code'].value_counts().rename_axis('country_code').reset_index(name='localidades')
sns.barplot(data=local_por_pais, x='country_code', y='localidades', palette="Set3")
plt.title("Cantidad de localidades por país")
plt.xlabel("País")
plt.ylabel("Cantidad de localidades")
plt.show()
```

## 🏆 Top 10 provincias con más localidades

```python
top_provincias = cities['state_name'].value_counts().head(10)
top_provincias.sort_values().plot(kind='barh', color='coral')
plt.title("Top 10 provincias con más localidades")
plt.xlabel("Cantidad de localidades")
plt.ylabel("Provincia")
plt.show()
```

## 📌 Conclusiones
- Brasil lidera en cantidad de provincias y localidades.
- Algunas provincias tienen una gran densidad urbana.
