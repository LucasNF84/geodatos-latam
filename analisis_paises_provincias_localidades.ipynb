# üåç An√°lisis de Pa√≠ses, Provincias y Localidades

Este notebook muestra un an√°lisis exploratorio de los datos de pa√≠ses, provincias y localidades de Am√©rica del Sur, utilizando los archivos descargados desde el repositorio `dr5hn/countries-states-cities-database`. Nos enfocamos en los siguientes pa√≠ses:

- üáßüá∑ Brasil
- üáµüáæ Paraguay
- üá∫üáæ Uruguay
- üáßüá¥ Bolivia
- üá®üá± Chile
- üáµüá™ Per√∫


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n est√©tica
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Leer CSVs
countries = pd.read_csv("countries.csv")
states = pd.read_csv("states.csv")
cities = pd.read_csv("cities.csv")
```

```python
# Filtrar solo los pa√≠ses deseados
target_codes = ['BR', 'PY', 'UY', 'BO', 'CL', 'PE']
countries = countries[countries['iso2'].isin(target_codes)]
states = states[states['country_code'].isin(target_codes)]
cities = cities[cities['country_code'].isin(target_codes)]
```

## üóÇÔ∏è Provincias por pa√≠s

```python
prov_por_pais = states['country_code'].value_counts().rename_axis('country_code').reset_index(name='provincias')
sns.barplot(data=prov_por_pais, x='country_code', y='provincias', palette="Set2")
plt.title("Cantidad de provincias por pa√≠s")
plt.xlabel("Pa√≠s")
plt.ylabel("Cantidad de provincias")
plt.show()
```

## üèôÔ∏è Localidades por pa√≠s

```python
local_por_pais = cities['country_code'].value_counts().rename_axis('country_code').reset_index(name='localidades')
sns.barplot(data=local_por_pais, x='country_code', y='localidades', palette="Set3")
plt.title("Cantidad de localidades por pa√≠s")
plt.xlabel("Pa√≠s")
plt.ylabel("Cantidad de localidades")
plt.show()
```

## üèÜ Top 10 provincias con m√°s localidades

```python
top_provincias = cities['state_name'].value_counts().head(10)
top_provincias.sort_values().plot(kind='barh', color='coral')
plt.title("Top 10 provincias con m√°s localidades")
plt.xlabel("Cantidad de localidades")
plt.ylabel("Provincia")
plt.show()
```
